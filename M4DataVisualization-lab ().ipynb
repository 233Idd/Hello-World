```python
# print how many rows are there in the table named 'master'
QUERY=""" 
SELECT COUNT(*)
FROM master;""" 


# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


```python
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

Plot a histogram of `ConvertedComp.`


```python
QUERY = """
SELECT ConvertedComp FROM master
"""

df = pd.read_sql_query(QUERY,conn)

sns.displot(
data=df,
x="ConvertedComp",
kind="hist",
aspect=1.4,
bins=20
)
```

### Box Plots


Plot a box plot of `Age.`


```python
QUERY = """
SELECT Age FROM master
"""

df = pd.read_sql_query(QUERY,conn)

df.plot(kind='box', figsize=(10,10))
plt.title('Box plot of Age')
plt.ylabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY = """
SELECT Age, WorkWeekHrs FROM master
"""

df = pd.read_sql_query(QUERY,conn)
df.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(10,6), color='blue')
plt.title('Scatter plot of Age and WorkWeekHrs')
plt.xlabel ('Age')
plt.ylabel('WorkWeekHrs')
plt.show()
```

```python
df.plot(kind='scatter', x='WorkWeekHrs', y='Age', figsize=(10,6), color='darkblue')
plt.title('Scatter plot of Age and WorkWeekHrs')
plt.xlabel ('WorkWeekHrs')
plt.ylabel('Age')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
"""

df = pd.read_sql_query(QUERY,conn)

df.plot(kind='scatter', x='WorkWeekHrs', y='CodeRevHrs', figsize=(14,8), alpha=0.5, color='green', s='Age')
plt.title('Bubble plot of WorkWeekHrs and CodeRevHrs')
plt.xlabel ('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT DISTINCT(DatabaseDesireNextYear), COUNT(DatabaseDesireNextYear) AS Total_Number  FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear 
ORDER BY Total_Number DESC
"""
```

```python
df = pd.read_sql_query(QUERY,conn)
df
```

```python
df_pie=df.head()

labels = df_pie['DatabaseDesireNextYear']
sizes = df_pie['Total_Number']

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```


### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs FROM master
WHERE Age BETWEEN  30 AND 35
"""

df = pd.read_sql_query(QUERY,conn)
df.head()
```

```python
df.plot(x='Age', kind='bar', stacked=True,
        title='Median WorkWeekHrs and CodeRevHrs')
plt.show()
```

```python
df.groupby('Age').median().plot(kind='bar', stacked=True,
        title='Median WorkWeekHrs and CodeRevHrs')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY = """
SELECT Age, ConvertedComp FROM master
WHERE Age BETWEEN 45 AND 60
"""
df = pd.read_sql_query(QUERY,conn)

df.groupby('Age').median().plot(kind='line', 
        title='Median ConvertedComp for all ages from 45 to 60')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
QUERY = """
SELECT MainBranch, COUNT(MainBranch) as Number FROM master GROUP BY MainBranch
"""
df = pd.read_sql_query(QUERY,conn)

df.head()
```

```python
# Create horizontal bars
plt.barh(y=df.MainBranch, width=df.Number)
plt.title('Bar Chart of Mainbranch')
plt.xlabel ('Number')
plt.ylabel('MainBranch')
```
